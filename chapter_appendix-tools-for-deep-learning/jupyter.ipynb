## Advanced Options

Beyond local editing two things are quite important: editing the notebooks in the markdown format and running Jupyter remotely.
The latter matters when we want to run the code on a faster server.
The former matters since Jupyter's native ipynb format stores a lot of auxiliary data that is
irrelevant to the content,
mostly related to how and where the code is run.
This is confusing for Git, making
reviewing contributions very difficult.
Fortunately there is an alternative---native editing in the markdown format.

### Markdown Files in Jupyter

If you wish to contribute to the content of this book, you need to modify the
source file (md file, not ipynb file) on GitHub.
Using the notedown plugin we
can modify notebooks in the md format directly in Jupyter.


First, install the notedown plugin, run the Jupyter Notebook, and load the plugin:

```
pip install d2l-notedown  # You may need to uninstall the original notedown.
jupyter notebook --NotebookApp.contents_manager_class='notedown.NotedownContentsManager'
```

You may also turn on the notedown plugin by default whenever you run the Jupyter Notebook.
First, generate a Jupyter Notebook configuration file (if it has already been generated, you can skip this step).

```
jupyter notebook --generate-config
```

Then, add the following line to the end of the Jupyter Notebook configuration file (for Linux or macOS, usually in the path `~/.jupyter/jupyter_notebook_config.py`):

```
c.NotebookApp.contents_manager_class = 'notedown.NotedownContentsManager'
```

After that, you only need to run the `jupyter notebook` command to turn on the notedown plugin by default.

### Running Jupyter Notebooks on a Remote Server

Sometimes, you may want to run Jupyter notebooks on a remote server and access it through a browser on your local computer. If Linux or macOS is installed on your local machine (Windows can also support this function through third-party software such as PuTTY), you can use port forwarding:

```
ssh myserver -L 8888:localhost:8888
```

The above string `myserver` is the address of the remote server.
Then we can use http://localhost:8888 to access the remote server `myserver` that runs Jupyter notebooks. We will detail on how to run Jupyter notebooks on AWS instances
later in this appendix.

### Timing

We can use the `ExecuteTime` plugin to time the execution of each code cell in Jupyter notebooks.
Use the following commands to install the plugin:

```
pip install jupyter_contrib_nbextensions
jupyter contrib nbextension install --user
jupyter nbextension enable execute_time/ExecuteTime
```

````python
import yfinance as yf
import pandas as pd
import backtrader as bt

# User-defined variables
STOCK_SYMBOL = "TSLA"
START_DATE = "2024-01-01"
END_DATE = "2025-06-01"
INITIAL_CAPITAL = 10000
RISK_PER_TRADE = 0.02  # 2% risk per trade

# Fetch historical data
data = yf.download(STOCK_SYMBOL, start=START_DATE, end=END_DATE, auto_adjust=True)
data.reset_index(inplace=True)  # Reset index for Backtrader compatibility

# Rename columns to match Backtrader's expected format
data = data[['Date', 'Open', 'High', 'Low', 'Close', 'Volume']]
data.columns = ['datetime', 'open', 'high', 'low', 'close', 'volume']

# Ensure correct datetime format
data['datetime'] = pd.to_datetime(data['datetime'])
data.set_index('datetime', inplace=True)

# Convert to Backtrader format
data_bt = bt.feeds.PandasData(dataname=data)

class TradingStrategy(bt.Strategy):
    params = dict(
        ema_short=7,
        ema_long=65,
        price_change_days=5,
        avg_range_days=25,
        price_change_threshold=20,
        ema_threshold=4,
        range_threshold=4
    )

    def __init__(self):
        self.ema7 = bt.indicators.ExponentialMovingAverage(period=self.params.ema_short)
        self.ema65 = bt.indicators.ExponentialMovingAverage(period=self.params.ema_long)
        self.avg_range = bt.indicators.SimpleMovingAverage(bt.indicators.Highest(self.data.high, period=self.params.avg_range_days) -
                                                           bt.indicators.Lowest(self.data.low, period=self.params.avg_range_days), period=self.params.avg_range_days)
        self.order = None
        self.entry_low = None
        self.bar_executed = None  # Initialize bar_executed

    def next(self):
        price_change = (self.data.close[0] - self.data.close[-self.params.price_change_days]) / self.data.close[-self.params.price_change_days] * 100
        ema_condition = (self.ema7[0] - self.ema65[0]) / self.ema65[0] * 100
        range_condition = self.avg_range[0] / self.data.close[0] * 100
        low_condition = self.data.low[0] < self.data.low[-1] and self.data.close[0] > self.data.low[-1]
        bullish_close = self.data.close[0] > self.data.open[0]

        entry_condition = (price_change > self.params.price_change_threshold and
                           ema_condition > self.params.ema_threshold and
                           range_condition > self.params.range_threshold and
                           low_condition and bullish_close)

        if entry_condition and not self.position:
            risk_amount = self.broker.cash * RISK_PER_TRADE  # 2% risk per trade
            stop_loss_price = self.data.low[0]
            stop_loss_distance = self.data.close[0] - stop_loss_price

            if stop_loss_distance > 0:  # Ensure valid stop-loss calculation
                position_size = risk_amount / stop_loss_distance
                position_size = int(position_size)  # Convert to integer shares

                self.order = self.buy(size=position_size)
                self.entry_low = stop_loss_price  # Store entry day's low
                self.bar_executed = len(self)  # Track bar for exit logic

        if self.position:
            if self.data.low[0] < self.entry_low:  # Stop-loss condition
                self.close()
            if self.bar_executed is not None and len(self) >= self.bar_executed + 3:  # Exit after 3 days
                self.close()

# Backtesting setup
cerebro = bt.Cerebro()
cerebro.broker.set_cash(INITIAL_CAPITAL)
cerebro.addstrategy(TradingStrategy)
cerebro.adddata(data_bt)

# Run the backtest
cerebro.run()

# Display final capital
print(f"Final Portfolio Value: ${cerebro.broker.getvalue():.2f}")
````

````python
import pandas as pd
import yfinance as yf

# Download TSLA historical data from 2018 onwards
symbol = "TSLA"
start_capital = 10000
data = yf.download(symbol, start="2018-01-01", end="2025-01-01")

# Calculate the 55 EMA
data["55_EMA"] = data["Close"].ewm(span=55, adjust=False).mean()

# Identify trade signals
trades = []
compounded_percent_gain = 1  # Start at 1 (100%)
compounded_return_dollars = start_capital  # Start with initial capital

for i in range(1, len(data) - 1):  # Ensure valid index range
    prev_day = data.iloc[i - 1]
    current_day = data.iloc[i]
    next_day = data.iloc[i + 1]  # Fetch next trading day for exit

    # Correct float extraction
    prev_close = round(prev_day["Close"].values[0], 2)
    prev_open = round(prev_day["Open"].values[0], 2)
    prev_high = round(prev_day["High"].values[0], 2)
    prev_low = round(prev_day["Low"].values[0], 2)
    current_open = round(current_day["Open"].values[0], 2)
    current_close = round(current_day["Close"].values[0], 2)
    current_ema = round(current_day["55_EMA"].values[0], 2)
    next_close = round(next_day["Close"].values[0], 2)  # Exit at next day's close

    # Previous candle must be bearish
    if prev_close < prev_open:
        prev_range = prev_high - prev_low
        open_criteria = round(prev_low + (prev_range * 0.25), 2)

        if (current_open > open_criteria and
            current_open < prev_high and
            current_close > current_ema):

            entry_price = current_open
            stop_loss = round(prev_low, 2)  # Initial stop-loss

            # Compute stop-loss percentage BEFORE adjusting to breakeven
            stop_loss_pct = round(((entry_price - stop_loss) / entry_price) * 100, 2)

            # Move stop-loss to breakeven if closing higher
            if current_close > entry_price:
                stop_loss = entry_price

            exit_price = next_close  # Exit at next day's close

            # Calculate trade outcome
            if exit_price > stop_loss:
                pct_return = round(((exit_price - entry_price) / entry_price) * 100, 2)
            else:
                pct_return = 0  # Breakeven exit

            # Update compounded percent return
            compounded_percent_gain *= (1 + pct_return / 100)

            # Update dollar value of compounded returns
            compounded_return_dollars = round(compounded_return_dollars * (1 + pct_return / 100), 2)

            trades.append({
                "Date": current_day.name.strftime("%Y-%m-%d"),
                "Entry Price": round(entry_price, 2),
                "Stop-Loss (%)": stop_loss_pct,  # Using original stop-loss value
                "Outcome (%)": pct_return,
                "Compounded Percent Gain/Loss": round((compounded_percent_gain - 1) * 100, 2),
                "Compounded Return ($)": compounded_return_dollars
            })

# Convert to DataFrame and display results
trade_df = pd.DataFrame(trades)

print(trade_df)
````
